# Quantum search algorithm and BJ model

This notebook compile all the tools developped to perform simulation and produce graphs useful for the paper **Quantum Search Algorithm without Quantum Counting**.

The intial idea was inspired by the Bixton-Jornter model where the initial state $\ket{a}$ is coupled to a ladder of reservoir state $\ket{b_n}$ as shown in this figure:

This system can be summarized by the following Hamiltonian:
$$
H_{BJ} = \epsilon_a\ketbra{a}{a} + \sum_n \left[ \epsilon_n\ketbra{b_n}{b_n} + \beta (\ketbra{a}{b_n} + \ketbra{b_n}{a}) \right]
$$

The number of states in the resrvoir plays a big role in the dynamic of such a system and two main cases appears:
- 1) Infinite case

If the number of resrevoir state is infinite, an analytical solution can be found for the evolution of the probability amplitude. Especially for the probability amplitudde of the state $\ket{a}$ we have the following formula:
$$
a(t) = e^{-i\epsilon_a t}  \sum_{n=0}^{\infty}e^{-\gamma(t-n\tau)/2}\mathcal{L}_n(\gamma(t-n\tau))\Theta(t-n\tau)
$$

The behavior is govern by two main parameters, the decay rate $\gamma$ and the revival time $\tau$, where:
$$
\begin{cases}
\gamma = 2\pi\frac{\beta^2}{\Delta}\\
\tau = \frac{2\pi}{\Delta}
\end{cases}
$$
What's particularly intersting is that on the initial time period $(0, \tau)$, the probability amplitude obeys an exponential decay: $a(t) = e^{-\gamma t}$, an analogous behaviour as the Fano-Anderson model (continuum case), and following "Fermi's golden rule".
After a time $\tau$, we see a revival in the probability amplitude followed by an almost exponential decay weighted by the Laguerre polynomial ($\mathcal{L}_n$).

- 2) Finite case

When there is only a finite number $N$ fo reservoir states, the behaviour of this system is more difficult to characterize. First of all, there is no way of finding an analytical expression for $a(t)$, because it would require to solve the following secular equation, where $E_m$ are the eigenvalues of $H_BJ$:
$$
E_m - \epsilon_a = \beta^2\sum_n \frac{1}{E_m - \epsilon_n}
$$

The finite case still exhibit a similar bahaviour as the infinite case, with decay intersped by almost peridocal revival of period $\tau$.
We are here intersted in the initial decay (before the first revival) and in the conditions that we can impose on the parameters of the system in order for it to still behaves as an exponential decay.
We know that for large N, the system will behave like the infinte case, but under the condition that $\Delta$ is large enough: if it's too small the reservoir states are grouped together and Rabi-like oscillations appears.


- ### Quantum search

Inspired with this model, we want to build a quantum search algorithm. The key idea is to being able to "trap" the state of the circuit in a reservoir made of the targets states, to improve the stability of Grover's algo.
The famous Grover's algorithm can be seen as the trotterized version of the evolution under the following Hamiltonian:
$$
H_G = \ketbra{+}{+}^{\otimes n} + \ketbra{S}{S} 
$$
We modyfied this by adding ancillas qubits to mimic the dynamics of the BJ model:
$$
H = -\ketbra{+}{+}^{\otimes n} \otimes\ketbra{+}{+}^{\otimes r}   - \ketbra{S}{S} \otimes \sum_{k=0}^{R-1}E_k\ketbra{k}{k}
$$
We can see that better with the developped form:
$$
H = \sum_{k=0}^{R-1}E_k\ketbra{S,k}{S,k} + \left(1 - \frac{M}{N}\right) \ketbra{\perp,+_r}{\perp,+_r } + \frac{1}{N}\sqrt{\frac{M(N-M)}{R}} 
\sum_{k=0}^{R-1} (\ketbra{S,k}{\perp,+_r } + \ketbra{\perp,+_r}{S,k}) + \frac{M}{N}\ketbra{S,+_r}{S,+_r}
$$
where can see that: $\gamma = 2\pi\frac{M(N-M)}{R \Delta N^2 }= 2\pi\frac{\lambda (1-\lambda)}{R \Delta }$ and $\tau = \frac{2\pi}{\Delta}$.



**The goal of this notebook is to compile all the tools that allows us to simulate the dynamics of those different systems to understand its behiviors and produce figures for the paper.**

# 1) Bixon-Jornter model

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm
from scipy.special import laguerre
from math import factorial
from scipy.optimize import fsolve
from scipy import stats
from fitter import Fitter
import random
```

#### Simulation of the evolution:

To simulate the BJ model, we write functions to write the hamiltonian in a matrix form, and then to compute the evolution by numerically diagonalizing it. The probability of the initial state is then given by:
$$
P_a(t) = |\bra{a}e^{-iHt}\ket{a}|^2
$$

```python
## Hamiltonian in matrix form

def ham(N, spacing, coupling):
    """
    Construct the matrix representation of H_BJ with N background states (==> H is N+1 x N+1)
    H:
    if i = 0:
        H[i,i] = 0
    else:
        H[i,i] = E_0 + spacing*i = (i + 1/2 - N/2)*spacing
        H[0, i] = coupling
    """
    H = np.zeros((N+1,N+1)) # H filled with 0
    E = [(k + 1/2 - N/2)*spacing for k in range(N)] # get the energy distribution
    for i in range(1, N+1): 
        # fill H
        H[i,i] = E[i-1]
        H[0,i], H[i,0] = coupling, coupling
    assert len(H) == N+1, "H_bj don't have the right dimension"
    return H

def hamOverlap(N, spacing, coupling, trans):
    """
    Construct the pertubed BJ model, with the additional term |+><+| weighted by the parameter trans
    """
    H = ham(N, spacing, coupling)
    H_trans = trans*np.ones((N+1, N+1))
    for i in range(N+1):
        H_trans[i,i], H_trans[i,0], H_trans[0,i] = 0, 0, 0
        
    return H + H_trans

## compute the evolution of probability of initial state

def evolution(N, spacing, coupling, maxtime=10, npoints=100, plot=True):
    """
    return the time array and the corresponding probability evolution of P_a(t) under the BJ hamiltonian
    - N: number of reservoir states
    - maxtime: maximum time for wich the proba is computed
    - npoints: number of points where the proba is computed
    - plot: True or False, plot the probability versus time
    """
    # get the hamiltonian and time array
    H, t = ham(N, spacing, coupling), np.linspace(0, maxtime, npoints)
    # compute proba by numerically diagonalizing H:
    proba0 = np.array([
        abs((expm(-1j*H*tau)@([1] + [0]*N))[0])**2 for tau in t
    ])
    # plot options
    if plot:
        plt.plot(t, proba0, '--', c='r', label='simulation')
        plt.legend()
        plt.xlabel('time')
        plt.ylabel(r"$P_0(t)$")
        plt.grid()
        plt.show()
    return t, proba0

def evolutionT(N, spacing, coupling, trans, maxtime=10, npoints=100, plot=True):
    """
    return time array and proba corresponding for the pertubed BJ model (trans = pertubative constant)
    """
    H, t = hamOverlap(N, spacing, coupling, trans), np.linspace(0, maxtime, npoints)
    proba0 = np.array([
        abs((expm(-1j*H*tau)@([1] + [0]*N))[0])**2 for tau in t
    ])
    if plot:
        plt.plot(t, proba0, '--', c='r', label='simulation')
        plt.legend()
        plt.xlabel('time')
        plt.ylabel(r"$P_0(t)$")
        plt.grid()
        plt.show()
    return t, proba0

def evolutionBJ(spacing, coupling, E_a=1, maxtime=10, npoints=100, order=3, plot=False):
    """
    Compute the analytical evolution of infinite BJ model using Laguerre polynomial
    ATTENTION: in the original paper, teh laguerre polynomial are such that L_n(0) = 0. The Laguerre polynomial imported from scipy don't respect that property
    THis can xplain the weird behavioiur of this function
    """
    time = np.linspace(0, maxtime, npoints)
    gamma = 2*np.pi*coupling**2/spacing
    tau = 2*np.pi/spacing
    a = lambda x: sum([np.exp(-gamma*(x - n*tau)/2)*((laguerre(n))(gamma*(x - n*tau)))*np.heaviside(gamma*(x - n*tau), 0.5) for n in range(order)]) if x != 0 else 1
    proba = [abs(a(t))**2 for t in time]
    if plot:
        plt.plot(time, proba, label=f'Exact infinite BJ, r={round(spacing/coupling,2)}')
        plt.grid()
        plt.legend()
    return time, proba
```

```python
#EXAMPLES:how to use of the above functions:


# choice of parameters:
coupling = 0.3
spacing = 0.15

maxtime, npoints = 75, 500

# compute important constants:
tau, gamma = 2*np.pi/spacing, 2*np.pi*coupling**2/spacing
print("tau = ", tau) 
print("1/gamma = ", 1/gamma)

# BJ model, almost infinite (N large):
N = 100
t, p = evolution(N, spacing, coupling, maxtime=maxtime, npoints=npoints, plot=False) # get the evolution of P_a(t)
plt.plot(t, p, label=f'sim - BJ model (N={N})')

# BJ model finite case
N = 25
t, p = evolution(N, spacing, coupling, maxtime=maxtime, npoints=npoints, plot=False)
plt.plot(t, p, label=f'sim - BJ model (N={N})')

# pertubed BJ model
pert = 0.05
t, p = evolutionT(N, spacing, coupling, pert, maxtime=maxtime, npoints=npoints, plot=False)
plt.plot(t, p, label=f'sim - pert BJ model (N={N})')

# exponential decay
plt.plot(t, abs(np.exp(-t*gamma/2)**2), '--', label='exp decay (infinte case)')

#Bj model infinte
t, p = evolutionBJ(spacing, coupling, E_a=0, maxtime=maxtime, npoints=npoints, plot=False)
plt.plot(t, p, label='exact infinte BJ model')

plt.axvline(x=tau, label="tau", c='black')
plt.axvline(x=5/gamma, label=r"$1/\gamma$", c='b')

plt.legend()
plt.grid()
plt.xlabel('time')
plt.ylabel('p(t)')
plt.title(r"$\gamma = $" + f'{round(gamma,1)}' + r" - $\tau = $" + f'{round(tau,1)}')
plt.show()
```

On the above figure, we can witness the dynamics of the BJ model: initial exponential decay followed by a revival at time $\tau$.
The truncated (finite) version exhibit oscillations, and the pertubed one differs slightly.

- Now, the main problem is on how to **bound the residual ocillations** for the finite case.
First steps, we can compute the amplitude of those oscillations numerically. For that, the following function run a simulation of the evolution of $p_a(t)$ between a time $t_{min} \approx \frac{5}{\gamma}$  to not take into account the initial decay and $\tau$ to exclude the revival.

One approach that seems tempting is using pertubation theory, but it onky works in the limit where $\beta << \Delta$.
**Pertubative approach:**

```python
def compute_Ps_t_perturbative(N, epsilon, g, time_array):
    """
    Approximate P_s(t) using perturbation theory.

    Args:
        N: number of background states on each side (total 2N+1)
        epsilon: energy spacing of background states
        g: coupling constant
        time_array: 1D array of time values

    Returns:
        P_s: array of same shape as time_array, containing P_s(t)
    """
    P_s = np.zeros_like(time_array, dtype=np.float64)

    # Contribution from degenerate states (|0> and |s>)
    A_deg = (1/2) * (np.exp(-1j * g * time_array) + np.exp(1j * g * time_array))

    # Contribution from non-degenerate states (k ≠ 0)
    A_nondeg = np.zeros_like(time_array, dtype=np.complex128)
    for k in range(1, N + 1):
        Ek = k * epsilon
        coeff = g / (-Ek)
        A_nondeg += (np.abs(coeff)**2) * np.exp(-1j * Ek * time_array)
        A_nondeg += (np.abs(coeff)**2) * np.exp(1j * Ek * time_array)  # symmetric term for -k

    # Total amplitude and probability
    A_total = A_deg + A_nondeg
    P_s = np.abs(A_total)**2

    return P_s
```

```python
# use of the pertubative approach:

time = np.linspace(0, 100, 1000)

N = 80
epsilon = 0.5
g = 0.1

p = compute_Ps_t_perturbative(N, epsilon, g, time)
plt.plot(time, p/p[0],label="pertubative approach")

t, p = evolution(2*N+1, epsilon, g, maxtime=100, npoints=5000, plot=False)
plt.plot(t,p, label="BJ model")

plt.grid()
plt.legend()
```

We can see that the pertubative approach can approximate the behaviour on short period of times, but the condition that the pertubation must be small compared to the energy difference is too restrictive for our usage.

### Bounding residuals oscillations:

To bound the "residual" oscillations, we can first observe its behaviour using simulation using the function below:

```python
def boundOsc(time, proba, tau, pth = 0.01, coeff=0.99):
    """
    Return a bound on the RESIDUAL oscillations using numerical simulation.
    Given a time and proba obtained after the use of the evolution function, it:
        - wait for the proba to get as low as the pth to start measuring amplitude ==> to exclude initial decay
        - stop before coeff*tau because the revival would false the results
    the coeff parameter makes the function stop at coeff*tau to be sure that no revival is taken into account
    As defined earlier, tau = 2pi/Delta
    """
    # find t when p < p_th: go trough the proba array and record the time when p gets below the thershold p_th:
    index_start = 0
    while proba[index_start] > pth:
        index_start += 1
        if index_start >= len(time): # in the case that the proba never gets below the threshold, return 1
            return 1
    index_stop = int(coeff*tau*(len(time)-1)/time[-1]) # index of time point right before tau
    if len(proba[index_start:index_stop]) > 0: # check that there are values in between the two times (tmin and tmax)
        return max(proba[index_start:index_stop]) 
    else:
        return 1
```

```python
## EXAMPLE: how to use this function:
N = 50
max_time, n_points = 10, 500
g = 0.3
spacing = 0.1
tau = 2*np.pi/spacing

for N in [20, 40, 70]:
    t, p = evolution(N, spacing, g, maxtime=max_time, npoints=n_points, plot=False)
    bound = boundOsc(t, p, tau)

    plt.plot(t, p, label=f'N={N}')
    plt.plot([t[0], t[-1]], [bound, bound], '--', label=f'N={50}')

plt.grid()
plt.legend()
plt.ylim((0,1))
```

We can also observe the dependancy of those amplitudes on the different parameters of the model (coupling $\beta$, spacing $\delta$ and number of resrevoir states $N$).

```python
def bound(N, spacing, coupling, npoints=100, margin=0.9):
    """
    given parameters of the BJ model, compute the amplitude of the residual oscillatons
    (It just use compute the time and the probability evolution to feed it to the function boundOsc)
    """
    max_time = margin*2*np.pi/spacing
    t, p = evolution(N, spacing, coupling, maxtime=max_time, npoints=npoints, plot=False)
    return boundOsc(t, p, max_time)
```

```python
# example of usage: plot the max amplitude of the residual oscillations for different parameters:
coupling = 0.5
spacing = 0.1
N = np.arange(10, 120, 5)


for spacing in [0.15, 0.2, 0.3, 0.4, 0.5]:
    bounds = [bound(n, spacing, coupling) for n in N]
    plt.plot(N, bounds, '--o', label=f'spacing = {spacing}')


plt.grid()
plt.legend()
plt.xlabel('number of states')
plt.ylabel('bound on proba oscillations')
plt.title(f'g={coupling}')
```

As expected, **the amplitude of those residuals oscillations diminish as the number of states $N$ gets bigger**. What's more, the dependancy on the spacing $\Delta$ confirms our first intuition: if the spacing gets too small, the system can't be really considered as a ladder.
We can also observe teh dependancy of the coupling parameter $\beta$

```python
coupling = 0.1
spacing = 0.2
N = np.arange(10, 120, 5)


for coupling in [0.1, 0.2, 0.3, 0.5, 0.7]:
    bounds = [bound(n, spacing, coupling) for n in N]
    plt.plot(N, bounds, '--o', label=r"$\beta = $"+f'{coupling}')


plt.grid()
plt.legend()
plt.xlabel('number of states')
plt.ylabel('bound on proba oscillations')
plt.title(f'g={coupling}')
```

#### - Deriving a bound with Probablisitic approach

In order to make sure that the scheme we derived is efficient, we need to make sure that the residuals can be controlled.
We know that the probability can be expressed from teh eigenvalues and eigenvectors of $H$ as:
$$
P_a(t) = |\sum_n |\braket{a|E_n'}|^2e^{-iE_n't}|^2
$$

In the infinite BJ model, we know that the distribtution of the eigenvectors on the intial state $\ket{a}$ can be expressed as a Lorentzian:
$$
p(E_n') = |\braket{a|E_n'}|^2 = \frac{\beta^2}{(E_n' - E_a)^2 + \beta^2 + (\beta^2\pi/\Delta)^2}
$$

What's more, the eigenvalues are located in precise intervals : $E_n' \in [(n-1)\Delta, n\Delta]$.

To estimate $P_a(t)$ behavior, we can try the following statistical approach:
- Draw $E_n'$ at random: $E_n' \sim \mathcal{U}([(n-1)\Delta, n\Delta])$
- Get the eigenvectors ditribution $\hat{p_n} = p(E_n') + w$ where $w$ is a noise to be detremined/fitted to the model.


Proposals:
- $w \sim \mathcal{N}(0, \sigma^2)$ follows a gaussian with a variance?
- $w \sim \mathcal{U}(-\delta, \delta)$

This gives us a random variable $Z_N = |\sum_n (p(E_n') + w)e^{-iE_n't}|^2$

```python
def lorentzian(E, coupling, spacing, Es=0):
    """
    lorentzian distribution of the eigenvector on initial state in infinite BJ model
    This compute p(E_n') as defined above
    """
    return coupling**2/((E - Es)**2 + coupling**2 + ((np.pi*coupling**2)/spacing)**2)

def evolutionRandEamp(time, N, delta, g, noiseAmp=0.1):
    """
    Compute the evolution of P_a(t) for:
    - eigenvalues E_n' drawn at random in [(n-1)D, nD]
    - eigenvecors amp |<a|E_n'>|**2 construced as p(E_n') + w where w is drawn at random in [-noiseAmp, noiseAmp]
    """
    # draw the eigenvalues at random:
    energies = [np.random.uniform((k - N/2)*spacing, (k + 1 - N/2)*spacing) for k in range(N)]
    # compute the coefficients p(E_n') + w
    amp = [lorentzian(E, g, delta, Es=0) + noiseAmp*(np.random.rand() - 0.5) for E in energies]
    s = []
    # compute the evloution P_a(t) as defined above
    for t in time: 
        Z = 0
        for i in range(len(energies)):
            Z += np.exp(-1j*energies[i]*t)*amp[i] 
        s.append(abs(Z)**2)
    return np.array(s)

def samplesPa(N, coupling, spacing, nsamples, noiseAmp=0.1):
    """
    Draw samples of the amplitude value of P_a(t) when the eigenvalues are drawn at random and P_a(t) computed as defined above (probabilistic approach)
    on the right time period (after initial decay and befire revival)
    - N, coupling, spacing: parameters of the BJ model
    - nsamples: number of samples
    """
    tmin, tmax = 5*spacing/(2*np.pi*coupling**2) , 2*np.pi/spacing
    time = np.arange(tmin, tmax, 0.1)
    samples = []
    for _ in range(nsamples):
        # draw the eigenvalues at random:
        energies = [np.random.uniform((k - N/2)*spacing, (k + 1 - N/2)*spacing) for k in range(N)]
        # compute the coefficients p(E_n') + w
        amp = [lorentzian(E, coupling, spacing, Es=0) + noiseAmp*(np.random.rand() - 0.5) for E in energies]
        # draw a time at random
        t = np.random.choice(time)
        # compute the amplitude and add it to the sample array
        samples.append(abs(sum([np.exp(-1j*energies[i]*t)*amp[i] for i in range(len(energies))]))**2)
    return samples
```

```python
# Random evolution compared to exact one:

# parameters:
N = 30
coupling = 0.3
spacing = 0.1
max_time = 2*np.pi/spacing
n_points = 1000

# exact BJ evolution
time, p = evolution(N, spacing, coupling, maxtime=max_time, npoints=n_points, plot=False)
plt.plot(time, p, '--', label=f'BJ model - N={N}')


# probabilistic approach:
noiseAmp = 0.1 # arbitrary value

nsamples = 100 # number of random eigenvalues drawing
p = evolutionRandEamp(time, N, spacing, coupling, noiseAmp=noiseAmp)
for _ in range(1,nsamples):
    p += evolutionRandEamp(time, N, spacing, coupling)    
p = p/nsamples

plt.plot(time, p/p[0], label=f'statistical - nsamples={nsamples}')

plt.grid()
plt.legend()
plt.show()
```

This approach seems to lead to reasonnable approximation on the oscillations: it seems that when taking the average of the evolution of $P^{(rand)}_a(t)$, we approximate the average value of the exact $P_a(t)$.  
Since $P_a$ is now a random variable we can try to **determine its probability law**. To do that, the code below plot an histogram of the amplitude of $P^{(rand)}_a(t)$ (frequency vs amplitude) for value of $t$ between $t_{min}$ and $t_{max}$ defined to avoid the initial decay and the revival.

```python
# create samples:

# params:
N = 30
coupling = 0.3
spacing = 0.1
nsamples = 100000
noiseAmp = 2*np.pi*coupling**2/(spacing*N)

# samples:
samples = samplesPa(N, coupling, spacing, nsamples, noiseAmp=noiseAmp)

#plot histogram:
nbins = 100
plt.hist(samples, density=True, bins=nbins, label='histogram')

# fit exponential law:
mean = np.array(samples).mean()
x = np.linspace(0, 1, nbins)
plt.plot(x, np.exp(-x/mean)/mean, '--', c='r', label='Exp law fit')

#plot details:
plt.grid()
plt.xlabel('amplitude')
plt.ylabel('frequency')
plt.legend()
```

From the observations, it seems like $Z_N \sim Exp(\lambda)$ where $E[Z_N] = 1/\lambda$.

In this case, $f_{Z_N}(x) = \lambda e^{-x\lambda}$.

We can then derive a bound: $P(Z_N > A) < p_{th} \Rightarrow A > -E[Z_N] \ln(p_{th}) $

**But** computing explicitly $E[Z_N]$ is no trivial and what's more, the theoritical justification of such an approximation is hard too. Even if we can't find an explicit formula, H is diagonalizable, meaning that its eigenvalues cannot be considered random. 

That's why a simpler approach on how to bound those oscillation has been also explored as shown below

### Another approach on the bounds:

The finite BJ model follows the following system of differential equations:
$$
\begin{align}
\frac{da}{dt} & = -i\epsilon_aa(t) -i\beta\sum_k b_n(t) \nonumber \\
\frac{db_n}{dt} & = -i\epsilon_k b_n(t) - i\beta a(t)
\end{align}
$$
We can set $\epsilon_a = 0$ and define $b_n = \hat{b_n}e^{-iE_nt}$, this gives $\frac{db_n}{dt} = \frac{d\hat{b_n}}{dt}e^{-E_nt} - iE_ne^{-E_nt}\hat{b_n}$ and we have:
$$
\frac{da}{dt} = -i\beta\sum_n\hat{b_n}e^{-iE_nt}
$$

We are trying to characterize the residual oscillations after decay ($a(t)$ reached 0), let's approximate $\hat{b_n} = \frac{1}{\sqrt{N}}$ so:
$$
\frac{da}{dt} = -i\beta\frac{\sin((N+1)\Delta t/2)}{\sin(\Delta t/2)} = -i\beta(1 + 2\sum_{k=1}^{N/2}\cos(k\Delta t))
$$
We can integrate and find:
$$
a(t) = -i \frac{2\beta}{\Delta\sqrt{N}}(\pi\frac{t}{\tau} + \sum_{k=1}^{N/2}\cos(2\pi k\Delta \frac{t}{\tau}))
$$

With this first approximation, we see that the amplitude seems to depend on the ratio $\theta = \frac{\beta^2}{R\Delta^2}$:
$$
A_{max} \propto \theta = \frac{\beta^2}{R\Delta^2}
$$

The code below test this assumption:

```python
# get the bounds for different value of the ration theta = beta²/delta²R   #R(delta^2/4beta^2)

# parameters
coupling = 0.1
theta = np.linspace(0.1, 1.1, 30)
Rarray = [10, 20, 30, 50, 70]

#compute the bounds
bounds = []
for R in Rarray:
    bounds.append([bound(R, coupling/np.sqrt(theta[i]*R), coupling) for i in range(len(theta))])
```

```python
# Plot the bounds computed above:
for k in range(len(Rarray)):
    plt.plot(theta, np.array(bounds[k]), '-', label=f'bounds,R={Rarray[k]}')

# Fit obtained by the approximation above:
coeff = 2
plt.plot(theta, coeff*theta, c='r', label=f'{coeff}'+r"$\theta$")

# format the plot:
plt.tick_params(axis='both', which='major',labelsize=20)
plt.legend()
plt.xlabel(r"$\theta = \frac{\beta^2}{R\Delta^2}$", fontsize=20)# , fontproperties=arial_font)
plt.ylabel(r"$A_{max}$", fontsize=20) #, fontproperties=arial_font)
plt.ylim((-0.05,1.05))

#comment/uncomment the following line to save the figure to the .eps format
#plt.savefig("Aboundvsr.eps", format='eps')

plt.show()
```

Above, we see that the approximation seems to be valid. For certains value of $\theta$ (between $0,0.3$?) the approximation that $A\propto \theta$ seems to be valid.

From that plot, it seems like $A_{max} \approx 1 - e^{-\alpha \theta}$?

To illustrate the above bound and the dynamics of the BJ model, we can have the followng plot:

# 2) Quantum Search

Now, we want to transfer those dynamics to the quantum search.
As said above, we came up with the following hamiltonian:
$$
H = \sum_{k=0}^{R-1}E_k\ketbra{S,k}{S,k} + \left(1 - \frac{M}{N}\right) \ketbra{\perp,+_r}{\perp,+_r } + \frac{1}{N}\sqrt{\frac{M(N-M)}{R}} 
\sum_{k=0}^{R-1} (\ketbra{S,k}{\perp,+_r } + \ketbra{\perp,+_r}{S,k}) + \frac{M}{N}\ketbra{S,+_r}{S,+_r}
$$
We can already identify this to the BJ model.

Then, we derive a quantum circuit from this by trotterization:
$$
e^{-iHt} \approx (e^{i\ketbra{+}{+}_d\otimes\ketbra{+}{+}_a\Delta t}e^{\ketbra{T}{T}_d\otimes H_a\Delta t})^n = (U_+U_T)^n
$$
where:
$$
\begin{cases}
U_+ = e^{i\ketbra{+}{+}_d\otimes\ketbra{+}{+}_a\Delta t} = I_d\otimes I_a - (1 - e^{i\Delta t})\ketbra{+}{+}_d\otimes\ketbra{+}{+}_a \\
U_T = e^{\ketbra{T}{T}_d\otimes H_a\Delta t} = \ketbra{\perp}{\perp}_d\otimes I_a + \ketbra{T}{T}\otimes e^{iH_a\Delta t}
\end{cases}
$$

In this part of the notebook, we build functions to simulate the evolution under the Hamiltonian and the circuit.

### Time evolution and dynamic of this system

```python
## TIME EVOLUTION:

#Here are the functions to simulate the time evolution of the state under this hamiltonian

def evolutionOperator(H, tmax, npoints):
    """
    Take an Hamiltonian H, a stopping time tmax and a number of points npoints and return two arrays, time and exp(-iHtime)
    """
    time = np.linspace(0, tmax, npoints)
    evolutionOp = [expm(-1j * tau * H) for tau in time]
    return time, evolutionOp


def getDataState(total_state, nqbits, nancillas):
    """
    Given a state total_state written in the computationnal basis of H_data x H_ancillas, the number of data and ancillas qbits,
    Return a kind of state collapsed onto the data state
    WARNING: not really proper but okay because just use as a tool to compute probabilty. CANT DO ANY OPERATION ON THE STATE THAT IS RETURNED
    ==> Better to use a density matrix in that case
    """
    data_state = []
    N, Na = 2**nqbits, 2**nancillas
    for i in range(N):
        data_state.append(np.sqrt(sum([abs(total_state[j])**2 for j in range(i * Na, (i+1)*Na)])))
    # check to see that evrything is ok: (norm = 1)
    norm = sum([abs(data_state[i])**2 for i in range(N)])
    if not(0.99 <= norm <= 1.1):
        print('norm ', norm)
        print('-------> data_state = ', data_state)
    assert 0.99 <= norm <= 1.1 # assert that the normalization is well respected (take into account small errors)
    return data_state


def Ha(nancillas, emin, emax, g=0.5, lin=True):
    """
    return the hamiltonian for ancillas qbit sassociated with targets states: diagonal matrix ha = diag(Ej)
    """
    N = 2**nancillas
    ha = np.zeros((N,N))
    if lin:
        energies = np.linspace(emin, emax, N)
    else:
        energies = np.random.uniform(emin, emax, size=(N,))
        
    for i in range(N):
        if i>0:
            ha[i,0], ha[0,i] = g, g
            ha[i,i] = energies[i]
            
    return np.diag(energies)
    

def Hd(nqbits, index_targets):
    """
    return the two data hamiltonians: Ht = |T><T| and H_plus = |+><+|
    """
    N, t = 2**nqbits, len(index_targets)
    Ht, H_plus = np.zeros((N, N)), (1/N) * np.ones((N,N)) # H_plus pretty straightforward
    # Construct Ht as |T><T|:
    state_T = [0] * N
    for i in index_targets:
        state_T[i] = np.sqrt(1/t)
    state_T = np.array(state_T)
    state_T = state_T.reshape(-1, 1)
    Ht = state_T @ state_T.conj().T
    
    return Ht, H_plus

def Htotal(nqbits, index_targets, nancillas, emin, emax):
    """
    Combine the differents hamiltonians to form H = -|+><+|x|+><+| - |T><T|xHa 
    """
    Nd, Na = 2**nqbits, 2**nancillas
    Ht, Hplus = Hd(nqbits, index_targets)
    
    return np.kron(Hplus, (1/Na)*np.ones((Na,Na))) + np.kron(Ht, Ha(nancillas, emin, emax))

def probafail_ancilla(nqbits, nancillas, index_targets, tmax, npoints, E1, E2, plot=True, gap=False):
    """
    Main function: given all parameters of the problem, compute the probaility of failure of the scheme for evry t between O and tmax
    Return time, proba_fa and can plot
    """
    assert max(index_targets) <= 2**nqbits # sanity check
    if gap == False: # just a way to input the energy level
        emin, emax = E1, E2
    else:
        emin, emax = E1 - E2/2, E1 + E2/2
    # get the global hamiltonian
    H = Htotal(nqbits, index_targets, nancillas, emin, emax)
    # compute the evolution operator
    time, evolutionOp = evolutionOperator(H, tmax, npoints)
    # intialize the state:
    state = np.sqrt(1/2**(nqbits + nancillas)) * np.array([1]*(2**(nqbits + nancillas)))
    # Get the indexes of the non-targets states (|ortho>)
    ortho_index = []
    for i in range(2**nqbits):
        if i not in index_targets:
            ortho_index.append(i)
    
    # Compute the proba: ie P_f(t) = sum(|<k,j|exp(-iHt)|+_tot>|² for k in ortho and j=0,..,Na-1) 
    proba_f = []
    for i in range(len(time)):
        s = evolutionOp[i] @ state
        s_data = getDataState(s, nqbits, nancillas)
        proba_f.append(sum([abs((s_data)[j])**2 for j in ortho_index]))
    
    # Plot options
    if plot:
        plt.plot(time, proba_f, label=f't={len(index_targets)},E={emin}-{emax},nA={nancillas},nD={nqbits}')
        plt.grid()
        plt.xlabel('time')
        plt.ylim(0,1)
        plt.legend()
        plt.ylabel('proba failure (not target state)')
    
    return time, proba_f

def probasuccess_ancilla(nqbits, nancillas, index_targets, tmax, npoints, E1, E2, plot=True, gap=False):
    """
    Main function: given all parameters of the problem, compute the probaility of failure of the scheme for evry t between O and tmax
    Return time, proba_fa and can plot
    """
    assert max(index_targets) <= 2**nqbits # sanity check
    if gap == False: # just a way to input the energy level
        emin, emax = E1, E2
    else:
        emin, emax = E1 - E2/2, E1 + E2/2
    # get the global hamiltonian
    H = Htotal(nqbits, index_targets, nancillas, emin, emax)
    # compute the evolution operator
    time, evolutionOp = evolutionOperator(H, tmax, npoints)
    # intialize the state:
    state = np.sqrt(1/2**(nqbits + nancillas)) * np.array([1]*(2**(nqbits + nancillas)))
    # Compute the proba: ie P_f(t) = sum(|<k,j|exp(-iHt)|+_tot>|² for k in Target and j=0,..,Na-1) 
    proba_s = []
    for i in range(len(time)):
        s = evolutionOp[i] @ state
        s_data = getDataState(s, nqbits, nancillas)
        proba_s.append(sum([abs((s_data)[j])**2 for j in index_targets]))
    
    # Plot options
    if plot:
        plt.plot(time, proba_s, label=f't={len(index_targets)},E={emin}-{emax},nA={nancillas},nD={nqbits}')
        plt.grid()
        plt.xlabel('time')
        plt.ylim(0,1)
        plt.legend()
        plt.ylabel('proba failure (not target state)')
    
    return time, proba_s

### we need to be able to compare that to the standard Grover's Behaviour (ie without ancillas qubits):

def probaSuccessGrover(nqbits, index_targets, tmax, npoints, plot=True):
    """
    Measure the proba of failure at a precise time (np.sqrt(N/p)*pi/2), and plot this vs the number of targets states
    ==> Use "standard Grover's algorithm" ie hamiltonian evolution without the ancillas qubits, and no TROTTER
    """
    N = 2**nqbits
    state = np.sqrt(1/2**(nqbits)) * np.array([1]*(2**(nqbits)))

    Ht, Hplus = Hd(nqbits, index_targets)
    H = Ht + Hplus
    # compute the evolution operator
    time, evolutionOp = evolutionOperator(H, tmax, npoints)
    # intialize the state:
    # Compute the proba: ie P_f(t) = sum(|<k,j|exp(-iHt)|+_tot>|² for k in Target and j=0,..,Na-1) 
    proba_s = []
    for i in range(len(time)):
        s = evolutionOp[i] @ state
        proba_s.append(sum([abs(s[j])**2 for j in index_targets]))
    
    # Plot options
    if plot:
        plt.plot(time, proba_s, label=f't={len(index_targets)},E={emin}-{emax},nA={nancillas},nD={nqbits}')
        plt.grid()
        plt.xlabel('time')
        plt.ylim(0,1)
        plt.legend()
        plt.ylabel('proba success - standard Grover')
    
    return time, proba_s
```

Above, we can see the evolution of the probability of success in time under $H$. We plotted $p_s(t) = \sum_{k} |\bra{T, k} e^{-iHt} \ket{\psi_0}|^2$

Naturally it is very similar to the BJ model, and behaves the same as the pertubed BJ model:
- initial approximately exponential decay, with a decay rate $\gamma =  \frac{\pi p(1-p)}{\Delta}$
- revival time around $t = \tau = \frac{2\pi}{\Delta}$
- residual oscillation between decay and revival coming from the finite reservoir
- oscillations in the decay, from the pertubed term

**Here, we can apply the bounds and the results from the BJ model to this**

Now, to derive a quantum circuit, we can trotterize this Hamiltonian.

```python
def is_unitary(A: np.array, th=1e-3) -> bool:
    test = A.dot(A.conj().T)
    return np.linalg.norm(test - np.identity(len(A))) < th

def are_equal(A: np.array, B: np.array, th=1e-3) -> bool:
    return np.linalg.norm(A - B) < th
```

```python
### TO CODE the circuit: instead of applying each gate individually, view evrything 
### as a big matrix that we are going to compute first, and then apply it to the state.
### Use object oriented-programming to keep track of everuthing when building the circuit

class QSAcircuit:
    """
    Class to represent and store the quantum circuit use for the quantum search. Each gate is stored using its matrix form,
    and the global circuit takes also the form of a matrix (combination of all the gates).
    Matrix form: M = (m_i,j), where m_i,j represent the coeff of |i><j| in the comp basis: 0<= i,j <= 2**(nqubits+nancillas +1) -1
    State represeted as a vector of length 2**(nqubits+nancillas +1), where each entry stores the coefficient of the basis state of the same index.
    """
    def __init__(self, nqubits, nancillas, index_targets, dt, emin, emax, p=1):
        """
        Create the circuit and initialize attributes
        """
        self.nqubits = nqubits
        self.nancillas = nancillas # attention, total number of ancillas qubits = nancillas + 1 (one more for oracle)
        self.N = 2**nqubits
        self.Na = 2**nancillas
        self.T = index_targets
        self.t = len(self.T)
        self.ortho = [i for i in range(self.N) if i not in self.T]
    
        self.qstate = np.array([])
        self.p = p
        
        self.tmeas = np.sqrt(self.N)*p
        self.dt = dt
        self.nsteps = int(self.tmeas/self.dt)
        self.energies = np.linspace(emin, emax, self.Na)
        self.emin, self.emax = emin, emax
        self.Ut = np.identity(2**(nqubits + nancillas + 1))
        self.Uplus = np.identity(2**(nqubits + nancillas + 1))
        self.Uiterate = np.identity(2**(nqubits + nancillas + 1))
        self.circuit = np.identity(2**(nqubits + nancillas + 1))
        
        self.probafail = 1
    
    ### HELPERS: function to create basic gates to help with the rest of the code
    
    def X(self):
        """
        Pauli X matrix
        """
        return np.array([[0, 1],
                         [1, 0]])
    
    def hadamard(self, n):
        """
        Hadamard gate on n qubits
        """
        h1 = np.sqrt(1/2) * np.array([[1, 1],
                                      [1, -1]])
        if n == 1:
            return h1
        else:
            H = np.identity(2)
            for i in range(n-1):
                H = np.kron(H, h1)
        
        assert len(H) == 2**n
        return H
    
    def one(self, i,j, n):
        A = np.zeros((n,n))
        A[i,j] = 1
        return A
        
    
    def oracle(self):
        """
        Compute the matrix representation of the oracle (O|j>|b> = |j>|b+1> if j in T, |j>|b> otherwise)
        """
        O_t = sum([np.kron(self.one(j,j, self.N), self.X()) for j in self.T])
        self.ortho = [i for i in range(self.N) if i not in self.T] #need to update ortho as well if T is getting updated
        O_ortho = sum([np.kron(self.one(j,j, self.N), np.identity(2)) for j in self.ortho])
        return O_t + O_ortho
    
    def c0H(self):
        """
        apply hadamard if control qubit in |0>
        """
        return np.kron(np.array([[1, 0],[0, 0]]), self.hadamard(self.nancillas)) +\
               np.kron(np.array([[0, 0],[0, 1]]), np.identity(self.Na))
    
    def c1H(self):
        """
        apply hadamard if control qubit |1>
        """
        return np.kron(np.array([[0, 0],[0, 1]]), self.hadamard(self.nancillas)) +\
               np.kron(np.array([[1, 0],[0, 0]]), np.identity(self.Na))
    
    def c0X(self):
        """
        apply X to +1 ancilla qubit if all the other qubits in state |0>, else nothing
        """
        d00, a00 = np.zeros((self.N,self.N)), np.zeros((self.Na, self.Na))
        d00[0,0], a00[0,0] = 1, 1
        controlled = np.kron(d00, np.kron(self.X(), a00))
        
        notcontrolled = np.kron(np.identity(self.N) - d00, np.kron(np.identity(2), (np.identity(self.Na) - a00)))
        
        return controlled + notcontrolled
    
    def phaseGate(self):
        """
        implement the phase gate for U_plus(dt)
        """
        return np.kron(np.identity(self.N), np.kron(np.array([[1, 0], [0, np.exp(1j*self.dt)]]), np.identity(self.Na)))
    
    def diagE(self):
        return np.diag([np.exp(1j*e*self.dt) for e in self.energies])
    
    def cExp(self):
        controlled = np.kron(np.array([[0, 0],[0, 1.0]]), self.diagE())
        uncontrolled = np.kron(np.array([[1.0, 0], [0, 0]]), np.identity(self.Na))
        return controlled + uncontrolled

    ### CREATION OF THE CIRCUIT AND ITS GATE ###
    
    def u_t(self):
        """
        Compute the matrix that represents the iterate U_T(dt)
        """
        H1 = np.kron(self.hadamard(self.nqubits), np.kron(np.identity(2), np.identity(self.Na)))
        O1 = np.kron(self.oracle(), np.identity(self.Na))
        C0H1 = np.kron(np.identity(self.N), self.c0H())
        expHA = np.kron(np.identity(self.N), np.kron(np.identity(2), self.diagE()))
        CexpHa = np.kron(np.identity(self.N) , self.cExp())
        C1H2 = np.kron(np.identity(self.N), self.c1H())
        #self.Ut =  H1 @ O1 @ C0H1 @ expHA @ C1H2 @ O1
        #self.Ut = O1 @ C0H1 @ expHA @ C1H2 @ O1
        self.Ut = O1 @ CexpHa @ O1
        return self.Ut
    
    def u_plus(self):
        """
        Compute the matrix that represents the iterate U_plus(dt)
        """
        H1 = np.kron(self.hadamard(self.nqubits), np.kron(np.identity(2), self.hadamard(self.nancillas)))
        C1 = self.c0X()
        phase = self.phaseGate()
        self.Uplus = H1 @ C1 @ phase @ C1 @ H1
        #v1 = H1 @ C1 @ phase @ C1 @ H1
        
        self.Uplus = np.kron(np.identity(self.N), np.kron(np.array([[1, 0], [0, np.exp(1j*self.dt)]]), np.identity(self.Na))) -\
                  (1 - np.exp(1j*self.dt))*np.kron((1/self.N)*np.ones((self.N, self.N)), np.kron(
                  np.array([[1, 0], [0, -1]]), (1/self.Na)*np.ones((self.Na, self.Na))))
        
        #print("TEST 2 METHODS:", are_equal(self.Uplus, v1))
        #print(np.linalg.norm(self.Uplus - v1))
        return self.Uplus
    
    
    def completeCircuit(self):
        """
        Compute the matrix reprsenting the complete circuit (ie iterate**nsteps)
        """
        ut, uplus = self.u_t(), self.u_plus()
        self.Uiterate = uplus @ ut
        self.nsteps = int(self.tmeas/self.dt)
        self.circuit = np.linalg.matrix_power(self.Uiterate, self.nsteps)
        #assert is_unitary(self.circuit)
        
    ## RUNNING THE CIRCUIT:
    
    def initializeQState(self, plus0=False):
        """
        Initialize the qstate to |+> x |0> x |0> (+ for data qubits, 0 for the oracle ancilla, and (0, .., 0) for teh others ancillas)
        """
        # ATTENTION, indexes: here we store the state in an array (c0, ..., c(nb_total_states_possible))
        # where the c(k) are the coefficient of the basis state |k> that links to the comp basis state as:
        # |k> = |q(0), q(1), .., q(nd-1), q(nd), q(nd+1), ..., q(nd+nA)> with k = sum(2**j * q(j) for j = 0,...,nd+nA)
        # 
        # So, here here, the state |+> x |0> x |0> = sqrt(1/N) *( |0..00> x |0> x |0> + |0..01> x |0> x |0> + ... + |1...11> x |0> x |0>)
        # with the convention here, those corresponds to the state sqrt(1/N)*(|0> + |1> + ... + |N-1>)
        
        # create an array of the right size filled with zeros and add the coeffcient sqrt(1/N) for the states 0 to N-1
        
        if plus0:
            self.qstate = np.array([1/np.sqrt(self.N)]*self.N + [0]*(2**(self.nqubits + self.nancillas + 1) -  self.N))
        #self.qstate = np.array([1] + [0]*(2**(self.nqubits + self.nancillas+1) - 1))
        
        
        ## We also can try another first state: |+>|0>|+>, in term of binary, it means that everu superspoition
        # has the same weight but with the middle ancilla set to 0:
        else:
            index_list = []
            for k in range(2**(self.nqubits + self.nancillas)):
                binary = bin(k)[2:] # remove '0b'
                binary = '0'*(self.nqubits + self.nancillas - len(binary)) + binary # add missing zeros in front
                # add '0' in the middle:
                binary = binary[:self.nqubits] + '0' + binary[self.nqubits:]
                index_list.append(int(binary, 2))
            self.qstate = np.array([1/np.sqrt(self.N * self.Na) if i in index_list else 0 for i in range(2**(self.nqubits + self.nancillas + 1))])
        #check that the extra ancilla is set to |0>:  
        assert self.checkExtraAncilla(), "extra ancilla is not set to |0> after initialization"
        assert (0.9999 <= np.sqrt(sum([abs(c)**2 for c in self.qstate])) <= 1.001), "Normalization of the state is not respected"
    
    def checkExtraAncilla(self, th=1e-10):
        """
        check that the extra ancilla (used for the oracle and the phase gate) is set to |0>.
        To do that, get the index of the global basis states (|k> for k in 2NNa) for wich the extra ancilla is 1 and check that 
        the coefficient associated is 0.
        In practice, we allow a small margin of error (th: threshold)
        """
        for k in range(2*self.N*self.Na):
            binary = bin(k)[2:] # to remove the '0b' in the binary conversion
            binary = '0'*(self.nqubits + self.nancillas + 1 - len(binary)) + binary
            extra_ancilla = binary[self.nqubits]
            if extra_ancilla == '1':
                if abs(self.qstate[k]) > th:
                    print('index: ', k)
                    print('binary: ', binary)
                    print('coeff :', abs(self.qstate[k]))
                    #print("STATE: ")
                    #print(self.qstate)
                    return False
            else:
                if abs(self.qstate[k]) < th:
                    return False
        return True
        
        
    def applyCircuit(self, info=False):
        """
        apply the complete circuit to the current state, and update the state of the circuit
        """
        self.initializeQState() # initialize the state
        self.completeCircuit() # compute the circuit and store it into self.circuit
        if info:
            print('### INFO: circuit matrix and inital state ###')
            print('circuit: ', self.circuit)
            print('size of circuit: ', len(self.circuit))
            print('--------------')
            print('intial state: ', self.qstate)
            print('size state: ', len(self.qstate))
            print('--------------')
        self.qstate = self.circuit @ self.qstate
        
        assert self.checkExtraAncilla(), f"extra ancilla is not set to |0> after running the circuit: t,E,nA,nD,tmeas: {self.T},{self.emin}-{self.emax},{self.nancillas},{self.nqubits},{round(self.tmeas,2)}"
        assert (0.9999 <= np.sqrt(sum([abs(c)**2 for c in self.qstate])) <= 1.001), "Normalization of the state is not respected"
    
    def probaFailure(self):
        """
        Compute the probabilty of failure if current state is measured
        """
        # First need to identify the indexes of the target states in the big vector, ie isolate the data qubits:
        # |k> = |q(0), q(1), .., q(nd-1), q(nd), q(nd+1), ..., q(nd+nA)> with k = sum(2**j * q(j) for j = 0,...,nd+nA)
        # in those total states we only want the "data state" that is |p> = |q(0), ... , q(nd-1)>, 
        # And the ones that are NOT in T ({target states}) to compute proba of failure (1 - Psuccess)
        index_total_state_ortho = []
        for k in range(2**(self.nqubits + self.nancillas + 1)):
            binary = bin(k)[2:] # to remove the '0b'
            binary = '0'*(self.nqubits + self.nancillas + 1 - len(binary)) + binary
        
            # ATTENTION, here the order is reversed: 0000...001 <-> 1 for example (and not 100...000 as in the description above)
            #binary = binary[::-1] # flip teh string to get it with the right convention
            #index_data = int(binary[self.nqubits:], 2)
            index_data = int(binary[:self.nqubits], 2)            
            if index_data not in self.T:
                index_total_state_ortho.append(k)
            #else:
            #    print(index_data, binary[:self.nqubits], k, binary)
                
        # Now, we have the list of the indexes that we need to take into account to compute the proba.
        # Let's run the circuit:
        self.applyCircuit()
        # and compute the proba of failure:
        probaF = sum([abs(self.qstate[j])**2 for j in index_total_state_ortho])
        self.probafail = probaF
        if probaF >= 1.1:
            print("WARNING, probabilty value too high: pF = ", probaF)
        return probaF  
    
    def checkProbaFailurEvol(self, tmax, npoints, plot=True, th=1e-5):
        """
        Allows to visualize the evolution of the failure proba along time of evolution, 
        by running the circuit with different time of measurement, ie incresingly higher number of steps
        """
        # creat list of time after measurement to see the evolution, init the proba fail list
        time, probaF_list = np.linspace(0, tmax, npoints), []
        for tau in time:
            self.tmeas = tau # update tmeas
            pf = self.probaFailure()
            if tau == 0:
                assert (pf - 1 + (len(self.T)/2**(self.nqubits))) <= th, "Without any action of the circuit, proba doesn't add up"
            probaF_list.append(pf) # compute the new failure probability
        
        # Plot option
        if plot:
            plt.plot(time, probaF_list, label=f'CIRCUIT - t,E,nA,nD: {len(self.T)},{self.emin}-{self.emax},{self.nancillas},{self.nqubits}')
            plt.grid()
        
        return probaF_list

    def checkProbaSuccessEvol(self, tmax, npoints, plot=True, th=1e-5):
        """
        Allows to visualize the evolution of the SUCCESS proba along time of evolution, 
        by running the circuit with different time of measurement, ie incresingly higher number of steps
        """
        # creat list of time after measurement to see the evolution, init the proba fail list
        time, probaF_list = np.linspace(0, tmax, npoints), []
        for tau in time:
            self.tmeas = tau # update tmeas
            pf = 1 - self.probaFailure()
            #if tau == 0:
                #assert (pf - 1 + (len(self.T)/2**(self.nqubits))) <= th, "Without any action of the circuit, proba doesn't add up"
            probaF_list.append(pf) # compute the new failure probability
        
        # Plot option
        if plot:
            plt.plot(time, probaF_list, label=f'CIRCUIT - t,E,nA,nD: {len(self.T)},{self.emin}-{self.emax},{self.nancillas},{self.nqubits}')
            plt.grid()
        
        return probaF_list
    
    def pFvsNbTarg(self, nmoy=1, plot=True, rd=False):
        """
        Compute the probability of failure after nsteps iteration (fixed by tmeas parameter) for every number of target states possible
        """
        self.T, probaF_list = [], []
        for i in range(1, self.N):
            # get a list of i target states (random)
            pr = 0
            if rd:
                for _ in range(nmoy):
                    self.T = random.sample(range(self.N), i)
            else:
                self.T.append(i-1)
            pr += self.probaFailure()
            probaF_list.append(pr/nmoy) # compute the new failure probability
        
        # Plot option
        if plot:
            plt.plot(np.arange(1, len(probaF_list)+1, 1), probaF_list, '-o', label=f'CIRCUIT - E,nA,nD,tmeas: {self.emin}-{self.emax},{self.nancillas},{self.nqubits},{round(self.tmeas,2)}')
            plt.grid()
        
        return probaF_list  
    
    def pFvsNbTargAdapted(self, nmoy=1, plot=True, rd=False):
        """
        Compute the probability of failure after nsteps iteration (fixed by tmeas parameter) for every number of target states possible 
        while ADAPTING the spacing parameter depending on the proportion of marked states.
        """
        self.T, probaF_list, meas_time = [], [], []
        for i in range(1, self.N):
            # get a list of i target states (random)
            pr = 0
            if rd:
                for _ in range(nmoy):
                    self.T = random.sample(range(self.N), i)
            else:
                self.T.append(i-1)
            if i > 1:
                spacing = abs(self.energies[1] - self.energies[0])*np.sqrt((i*(1 - i/self.N))/((i-1)*(1 - (i-1)/N))) #get the new spacing from the previous one
                emin, emax = 1 - spacing*self.Na/2, 1+ spacing*self.Na/2
                print(f'=> computing for M={i}, spacing = {spacing}, emin={emin},emax={emax}')
                self.energies = np.linspace(emin, emax, self.Na)
                coupling = np.sqrt(i*(self.N - i)/self.Na)/N
                print(f'coupling={coupling}')
                gamma = 2*np.pi*coupling**2/spacing
                print(1/gamma)
                self.tmeas = self.p / gamma
                print(f'tmeas = {self.tmeas} - tau = {2*np.pi/spacing}')
                meas_time.append(int(self.tmeas/np.pi) + 1)
                print('d/b = ', spacing/coupling)
            pr += self.probaFailure()
            probaF_list.append(pr/nmoy) # compute the new failure probability
        
        # Plot option
        if plot:
            plt.plot(np.arange(1, len(probaF_list)+1, 1), probaF_list, '-o', label=f'CIRCUIT - E,nA,nD,tmeas: {self.emin}-{self.emax},{self.nancillas},{self.nqubits},{round(self.tmeas,2)}')
            plt.grid()
        
        return meas_time, probaF_list  
    
    
    def steps2min(self, tmax=50):
        """
        Compute the number of steps needed to achieve a minimal probability of failure for every number of target states possible
        - Loop over number of target states
        - compute evolution until tmax
        - compute minimum proba fail and minimal number of steps to get there
        """
        npoints = 2*tmax
        self.T, minN, minP = [], [], []
        for i in range(1, self.N):
            self.T = random.sample(range(self.N), i)
            pf = np.array(self.checkProbaFailurEvol(tmax, npoints, plot=False))
            minP.append(min(pf))
            minN.append(np.argmin(minP))

        index_targets = list(range(1, self.N))
        fig, ax1 = plt.subplots()
        ax1.plot(index_targets, minN, '-o', color='r', label='number of steps')
        ax1.legend()
        ax1.set_xlabel('number of target states')
        ax1.set_ylabel('Number of steps to get to the proba')
        ax2 = ax1.twinx()
        ax2.plot(index_targets, minP, '-o')
        ax2.set_ylabel("pf")
        plt.grid()
        
```

```python
def evolOp_trot(H1, H2, tf, trot_param, fixed='nsteps'):
    """
    Compute the trotterized evolution operator of the Hamiltonian H = H1 + H2 ==> exp(-iHtf)
    Parameters:
     - fixed: either 'nsteps' (fix nsteps for trot) or 'dt' (fix dt for trot) ==> trot_param will be attribited based on that
    """
    assert fixed == 'nsteps' or fixed == 'dt'
    if fixed == 'nsteps':
        nsteps = trot_param
        dt = tf/nsteps
    else:
        dt = trot_param
        nsteps = int(tf/dt)
        
    U1, U2 = expm(-1j * H1 * dt), expm(-1j * H2 * dt)
    return np.linalg.matrix_power(U1 @ U2, nsteps)


def probaSuccess_trotSG(nqbits, index_targets, tmax, npoints, trot_param, fixed='dt'):
    """
    compute evolution of trotterized (circuit version) of grover algorithm
    """
    assert max(index_targets) <= 2**nqbits # sanity check
    Na = 2**nancillas
    
    # get the hamiltonians H1 and H2
    H1, H2 = Hd(nqbits, index_targets)

    # compute the TROTTERIZED evolution operator
    time = np.linspace(0, tmax, npoints)
    evolutionOp = [evolOp_trot(H1, H2, tf, trot_param, fixed=fixed) for tf in time]
    # intialize the state:
    state = np.sqrt(1/2**(nqbits)) * np.array([1]*(2**(nqbits)))
   
    # Get the indexes of the non-targets states (|ortho>)
    # Compute the proba: ie P_f(t) = sum(|<k,j|exp(-iHt)|+_tot>|² for k in ortho and j=0,..,Na-1) 
    proba_f = []
    for i in range(len(time)):
        s = evolutionOp[i] @ state
        proba_f.append(sum([abs(s[j])**2 for j in index_targets]))
    
    return time, proba_f
```

## Efficiency on multiple proportions of targets states

Rigth now, how to choose the optimal parameters of the circuit?

- 1) Enough time between the intiial decay and the revival time:
 
The decay is given by $1/\gamma$ and the revival time is $\tau$. Therefore, choosing $\frac{\tau}{1/\gamma} = \tau\gamma = \alpha$ should allow to get the probability as high as $1 - e^{-\alpha}$. This gives the following condition:
$$
\frac{\Delta}{\beta} \leq \frac{2\pi}{\sqrt{\alpha}}
$$

- 2) Bound on the reisudal oscillation
 
The finite nature of the reservoir induce residual oscillations. The maximum amplitude $A$ of those oscillations can be bounded by:
$$
A \leq 2\frac{\beta^2}{\Delta^2R}
$$

- Parameters choice
Using those bounds, we can then choose the parameters:
$$ 
R \geq \frac{\alpha}{2\pi^2A}
$$
$$
\frac{\sqrt{\alpha}}{\pi A}\frac{\beta}{R} \leq \Delta \leq \frac{2\pi}{\sqrt{\alpha}}\beta
$$

**Choosing $\alpha$ and $A$ determine completely the efficiency of the model: we can compute the success probability $P_s$ and the scaling $T$.**
$$
P_s \geq 1 - (e^{-\alpha} + A)
$$
$$
T = \frac{\sqrt{\alpha}}{\pi^2A}\frac{1}{\sqrt{R}}\frac{N}{\sqrt{M(N-M)}}
$$

**QUESTION**: here we saw that $\Delta \propto \beta$ in order to get those performance, and $\beta = \frac{\sqrt{M(N-M)}}{N\sqrt{R}} = \sqrt{\frac{\lambda(1-\lambda)}{R}}$ depends directly on the proportion of marked states $\lambda$. Is it possible to choose $\Delta$ such that the scheme is efficeient on a large proportion of markes states?

- Try to set the parameters for the worst case ($\lambda=1/N$)

Here, we chose $R = \frac{\alpha}{2\pi^2A}$ and $\Delta = \frac{\sqrt{\alpha}}{\pi A}\frac{\beta(\lambda=1/N)}{R}$


First bound on the time between the decay and the revival gives:
$$
\frac{\Delta}{\beta} = \sqrt{\frac{1/N(1 - 1/N)}{\lambda(1 - \lambda)}}\frac{2\pi}{\sqrt{\alpha}}
$$

```python
l = np.linspace(0.001, 0.999, 1000)
for N in [5, 10, 50, 100, 1000]:
    plt.plot(l, np.sqrt(((1 - 1/N)/N)/(l*(1-l))), label=f'{N}')

plt.plot(l, [1]*len(l), '--', c='b')
plt.grid()
plt.legend()
plt.xlabel(r"$\lambda$")
```

As shown on the figure above, if N is big enough we can still consider that $\frac{\Delta}{\beta} \leq \frac{2\pi}{\sqrt{\alpha}}$.

2nd bound: bounding of the amplitudes:
$$
2\frac{\beta^2}{\Delta^2R} = \frac{\lambda(1 - \lambda)}{1/N(1 - 1/N)}A 
$$

The question is $\frac{\lambda(1 - \lambda)}{1/N(1 - 1/N)} \geq 1?$

```python
l = np.linspace(0.001, 0.999, 1000)
for N in [5, 10, 50, 100]:
    plt.plot(l, (l*(1-l))/((1 - 1/N)/N), label=f'{N}')

plt.plot(l, [1]*len(l), '--', c='b')
plt.grid()
plt.legend()
plt.xlabel(r"$\lambda$")
```

This choice of $\Delta$ yields to a scaling:
$$
T \propto \frac{2\sqrt{\alpha}}{A}\frac{1}{\sqrt{R}}\sqrt{N-1}\frac{N}{M(N-M)} \approx \sqrt{N}?
$$

This implies that this scheme should be efficient on a large number of proportion of marked states, even more for $R \rightarrow \infty$.

Below, we can observe the probability of success as a function of the number of marked states for different parameters $\alpha, A$ when $R,\Delta$ are fixed according to the above.

```python
def parameters(alpha, A, nqubits, tch=2, coeff=15):
    R =  alpha/(2*A*np.pi**2)
    print('R brut = ', R)
    nancillas =  int(np.log2(R)) + 1
    print('nancillas=', nancillas)
    R = 2**nancillas
    print('R final = ', R)
    N = 2**nqubits
    print('N = ', N)
    tchoice =  tch
    spacing = 2*(np.sqrt(alpha)/(np.pi*A*R))*np.sqrt(tchoice*(1 - tchoice/N)/(N*R))
    print('spacing = ', spacing)
    delta = R*spacing/2
    print('delta = ', delta)
    print('tau = ', 2*np.pi/spacing)
    p =  coeff * (np.sqrt(alpha)/(A*np.sqrt(R)*2*np.pi**2) ) 
    print('tmeas = ', p*np.sqrt(N))
    gam = 2*np.pi/spacing
    print('1/gam_min = ',1/(gam*((1-1/N)/(N*R))), ' - 1/gam_max = ', 1/(gam*((1-1/2)/(2*R))))
    return nqubits, nancillas, delta, p
```

```python
nqubits = 5
N = 2**nqubits
alpha, A = [25], [0.3, 0.2]
tch = [2, 20]

circuits = []
probas = []

for i in range(len(A)):
    for j in range(len(alpha)):
        for k in range(len(tch)):
            nd, na, d, p = parameters(alpha[j], A[i], nqubits, tch=tch[k])
            c = QSAcircuit(nqubits, nancillas, [0], np.pi, 1 - d, 1 + d, p=p)
            circuits.append(c)
            print(' --------------------- computing A =',A[i]) 
            probas.append(c.pFvsNbTarg(plot=False))
            plt.plot(list(range(1, N)), 1 - np.array(probas[-1]), '--o', label=f'A={A[i]},alpah={alpha[j]},tch={tch[k]},p={round(p*N,2)}')

plt.grid()
plt.legend()
plt.ylim((0,1))
plt.ylabel(r"$P_S(t_m)$")
plt.xlabel('M/N')
```

```python
## PERFORMANCES WHEN SPACING IS ADAPTED:
nqubits = 5
N = 2**nqubits
alpha, A = 30, 0.1
tch = 1

nd, na, d, p = parameters(alpha, A, nqubits, tch=tch)
c = QSAcircuit(nd, na, [0], np.pi, 1 - d, 1 + d, p=5)
print(f' -- emin={1-d},emax = {1+d} -- ')
nsteps, probas = c.pFvsNbTargAdapted(plot=False)
plt.plot(list(range(1, N)), 1 - np.array(probas), '--o', label=f'A={A},alpah={alpha},tch={tch}') #,p={round(p*N,2)}')

plt.grid()
plt.legend()
plt.ylim((0,1))
plt.ylabel(r"$P_S(t_m)$")
plt.xlabel('M/N')
```

```python
l = np.linspace(2, N, N-2)
plt.plot(l, nsteps, '-o')
plt.plot(l, 5*np.sqrt(1/l), '-o')
plt.grid()
```
